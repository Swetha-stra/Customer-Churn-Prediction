```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='Churn')
plt.title("Churn Distribution")
plt.show()
```

```python
numeric_features = ['MonthlyCharges', 'TotalCharges', 'tenure']

# Convert TotalCharges to numeric (handle blanks)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Histograms
df[numeric_features].hist(figsize=(10,5))
plt.show()

# Boxplots
for col in numeric_features:
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
corr = df[numeric_features].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation between numeric features")
plt.show()
```

```python
sns.countplot(data=df, x='Contract', hue='Churn')
plt.title("Contract vs Churn")
plt.show()
```

```python
df['tenure_group'] = pd.cut(df['tenure'], bins=[0, 12, 24, 48, 60, 72], labels=['0-12','12-24','24-48','48-60','60-72'])
sns.countplot(data=df, x='tenure_group', hue='Churn')
plt.title("Tenure group vs Churn")
plt.show()
```
